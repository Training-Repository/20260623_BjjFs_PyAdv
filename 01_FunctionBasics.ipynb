```python
def Add(a, b, *others):     # Variable Args
    sum  = a + b
    print(others)
    for val in others:
        sum += val

    return sum


Add(1, 2), Add(1, 2, 3, 4), Add(1, 2, 3)
```

```python
def PrintName(ceo, cto, **others):      # Keatwoirded Variable list
    print(f"ceo = {ceo}")
    print(f"cto = {cto}")
    print(type(others), others)
    for position, person in others.items():
        print(f"{position} = {person}")


PrintName("Manish", "Abhijeet", cfo="Rakesh", cxo="Varun")
```

```python
def PrintEmp(id, name, city):
    print(f"{id = }")
    print(f"{name = }")
    print(f"{city = }")


def Logged_PrintEmp(*vArgs, **kwArgs):  # Universal Arg List
    print(f"vArg-count = {len(vArgs)}")
    print(f"kwArg-count = {len(kwArgs)}")

    print("Method called")
    PrintEmp(*vArgs, **kwArgs)
    print("Method returned")

Logged_PrintEmp(100, "Pravin", "Pune")
Logged_PrintEmp(100, city="Pune", name="Pravin")
```
